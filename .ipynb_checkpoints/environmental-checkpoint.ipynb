# Data Processing and Cleaning

We will focus on cleaning and processing the key weather measurements, specifically maximum temperature, minimum temperature, average temperature, precipitation, snowfall, and wind speed. These variables are crucial for integrating with wildfire data and building predictive models.

### Fill Temprature Value

We implemented a custom function to handle missing temperature values by performing linear interpolation and mean across the same day in different years. This method preserves day-specific trends while smoothing out natural fluctuations over the years.


```python
def fill_temperature(df, column, threshold_low=10, threshold_high=25):
    df['Year'] = df['DATE'].dt.year
    df['MonthDay'] = df['DATE'].dt.strftime('%m-%d')

   
    stations = df['STATION'].unique()
    for station in stations:
        station_data = df[df['STATION'] == station]
        missing_percentage = station_data[column].isna().mean() * 100
        
        # Interpolate for low missing data
        if missing_percentage < threshold_low:
            df.loc[df['STATION'] == station, column] = station_data[column].interpolate(method='linear', limit_direction='both')

        # Interpolate first, then adjust with monthly mean for moderate missing data
        elif missing_percentage < threshold_high:
            interpolated = station_data[column].interpolate(method='linear', limit_direction='both')
            large_gaps = interpolated.isna()
            df.loc[df['STATION'] == station, column] = interpolated.fillna(
                station_data.groupby(station_data['DATE'].dt.month)[column].transform('mean')
            )
          
        # Interpolate then fill with monthly mean for high missing data
        else:
            interpolated = station_data[column].interpolate(method='linear', limit_direction='both')
            large_gaps = interpolated.isna()
            df.loc[df['STATION'] == station, column] = interpolated.fillna(
                station_data.groupby(station_data['DATE'].dt.month)[column].transform('mean')
            )
 
    df.drop(columns=['Year', 'MonthDay'], inplace=True)
    
    return df

# Apply the function for TMAX
env_df = fill_temperature(env_df, 'TMAX')

env_df[['DATE', 'STATION', 'TMAX']].head(10)
```

```python
def check_temperature_across_years(env_df, station_id, month_day, column='TMAX'):
    # Filter the data for the specified station
    station_data = env_df[env_df['STATION'] == station_id].copy()
    station_data['YEAR'] = station_data['DATE'].dt.year
    
    station_data['MonthDay'] = station_data['DATE'].dt.strftime('%m-%d')
    specific_date_data = station_data[station_data['MonthDay'] == month_day].copy()
    specific_date_data = specific_date_data.sort_values('YEAR')
    
    # Plot the temperature across the years
    plt.figure(figsize=(10, 6))
    plt.plot(specific_date_data['YEAR'], specific_date_data[column], marker='o', linestyle='-')
    plt.title(f'{column} on {month_day} Across Years for Station {station_id}')
    plt.xlabel('Year')
    plt.ylabel(f'{column} (F)')
    plt.grid(True)
    plt.show()
    
    return specific_date_data[['YEAR', column]]

station_id = 'USC00047195'  
month_day = '01-01'        
temperature_across_years = check_temperature_across_years(env_df, station_id, month_day, 'TMAX')
```

### Adding TAVG (Average Temperature) Column
We will calculate and add the average temperature (TAVG) by averaging the maximum temperature (TMAX) and minimum temperature (TMIN) for each record.

```python
# Add TAVG (Average Temperature) column after TMIN
env_df['TAVG'] = (env_df['TMAX'] + env_df['TMIN']) / 2

# Rearrange columns
env_df = env_df[['STATION', 'NAME', 'LATITUDE', 'LONGITUDE', 'ELEVATION', 'DATE', 'TMIN', 'TMAX', 'TAVG', 'PRCP', 'SNOW', 'AWND']]
```

### Processing SNOW (Snowfall)
We will clean and process the SNOW data. 

```python
# Calculate the percentage of missing values in SNOW for each station
missing_snow_percentage = env_df.groupby('STATION')['SNOW'].apply(lambda x: x.isna().mean() * 100)
missing_snow_percentage
```

```python
env_df["SNOW"].isna().sum(), env_df["SNOW"].unique()[:20]
```

```python
def fill_snow_missing_data(df, column="SNOW", threshold_low=10, threshold_high=25):
    # Iterate over each station
    for station in df['STATION'].unique():
        station_data = df[df['STATION'] == station]
        missing_percentage = station_data[column].isna().mean() * 100
    
        if missing_percentage < threshold_low:
            # Interpolate for low missing data
            df.loc[df['STATION'] == station, column] = station_data[column].interpolate(method='linear', limit_direction='both')
        
        elif missing_percentage < threshold_high:
            # Interpolate first, then adjust with monthly mean for moderate missing data
            interpolated = station_data[column].interpolate(method='linear', limit_direction='both')
            large_gaps = interpolated.isna()
            if large_gaps.any():
                df.loc[df['STATION'] == station, column] = interpolated.fillna(
                    station_data.groupby(station_data['DATE'].dt.month)[column].transform('mean')
                )
            else:
                df.loc[df['STATION'] == station, column] = interpolated
        
        else:
            # Interpolate then fill with monthly mean for high missing data
            interpolated = station_data[column].interpolate(method='linear', limit_direction='both')
            df.loc[df['STATION'] == station, column] = interpolated.fillna(
                station_data.groupby(station_data['DATE'].dt.month)[column].transform('mean')
            )
    
    return df

env_df = fill_snow_missing_data(env_df, 'SNOW')

env_df["SNOW"].isna().sum(), env_df["SNOW"].unique()[:20]
```

```python
# Calculate the percentage of missing values in AWND for each station
missing_wind_percentage = env_df.groupby('STATION')['AWND'].apply(lambda x: x.isna().mean() * 100)
missing_wind_percentage
```

```python
# Drop the AWND (Average Wind Speed) column
env_df.drop(columns=['AWND'], inplace=True)
```

### Excluding Wind Speed Data (AWND)

Due to the complete lack of wind speed data (`AWND`) for most stations, and the potential bias of using data from a single station, we have decided to exclude wind speed from our wildfire prediction model. The focus will be on other environmental factors, such as temperature and precipitation, which provide a strong foundation for our predictive models.

## Standardizing Environmental Data Units 

```python
# Convert temperature from Fahrenheit to Celsius
env_df['TMAX'] = ((env_df['TMAX'] - 32) * 5.0/9.0).round(2)
env_df['TMIN'] = ((env_df['TMIN'] - 32) * 5.0/9.0).round(2)
env_df['TAVG'] = ((env_df['TAVG'] - 32) * 5.0/9.0).round(2)

# Convert PRCP and SNOW from inches to millimeters
env_df['PRCP'] = (env_df['PRCP'] * 25.4).round(2)
env_df['SNOW'] = (env_df['SNOW'] * 25.4).round(2)

# Rename columns to align with the wildfire dataset
env_df.rename(columns={
    'STATION': 'Station',
    'NAME': 'Name',
    'LATITUDE': 'Latitude',
    'LONGITUDE': 'Longitude',
    'ELEVATION': 'Elevation',
    'DATE': 'Date',
    'TMAX': 'MaxTemperature',
    'TMIN': 'MinTemperature',
    'TAVG': 'AvgTemperature',
    'PRCP': 'Precipitation',
    'SNOW': 'Snowfall',
}, inplace=True)

env_df.head()
```
