```python
import numpy as np
import pandas as pd
from pytrends.request import TrendReq
```

```python
df = pd.read_csv('cleaned_wilfire_environmental_data.csv')
```

```python
df.head()
```

```python
# Split data into pre-2004 and post-2004
pre_df = df[df['DiscoveryDate'] < '2004-01-01']
post_df = df[df['DiscoveryDate'] >= '2004-01-01']
pre_df
```

```python
post_df.reset_index(drop=True, inplace=True)
post_df
```

```python
df.columns
```

```python
# Define the start and end dates
start_date = pd.to_datetime('2004-01-01')
end_date = pd.to_datetime('2020-12-31')

timeframe_list = []

# Add 8-month intervals to get daily trend data
current_date = start_date
while current_date < end_date:
    next_date = current_date + pd.DateOffset(months=8) - pd.Timedelta(days=1)
    
    if next_date > end_date:
        next_date = end_date  
    
    timeframe_list.append(f'{current_date.strftime("%Y-%m-%d")} {next_date.strftime("%Y-%m-%d")}')
    current_date = next_date + pd.Timedelta(days=1)

timeframe_list
```

```python
# Initialize pytrends request
pytrends = TrendReq(hl="en-US", tz=360)
search_terms = ['wildfire', 'forest fire', 'california fire', 'fire season', 'fire prevention']


# fetch google trend data
all_trend_data = []

for timeframe in timeframe_list:
    pytrends.build_payload(search_terms, timeframe=timeframe, geo='US-CA')
    data = pytrends.interest_over_time()
    
    # drop inPartial column if exists
    if 'isPartial' in data.columns:
        data = data.drop(columns=['isPartial'])
        
        all_trend_data.append(data)

# Concatenate all data
trend_data = pd.concat(all_trend_data)
trend_data.reset_index(inplace=True)
trend_data
```

```python
trend_data['GoogleTrendAverage'] = trend_data[['wildfire','forest fire','california fire','fire season','fire prevention']].mean(axis=1)
trend_data
```

```python
# Merge google trend average and post 2004 wildfire data
post_df = post_df.copy()
trend_data['date'] = pd.to_datetime(trend_data['date'])
post_df['DiscoveryDate'] = pd.to_datetime(post_df['DiscoveryDate'])
merged_post_df = pd.merge(post_df, trend_data[['date', 'GoogleTrendAverage']], left_on='DiscoveryDate', right_on='date', how='left')
merged_post_df.drop(columns=['date'], inplace=True)
```
